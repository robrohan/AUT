```python
from pathlib import Path
from os.path import basename, splitext
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gpt_mini.midi_encoder as midichar
import csv
import math
import json
from gpt_mini.config import DEFAULT_DEVICE, CONFIG
```

```python
import logging
logging.basicConfig()
logging.getLogger().setLevel(logging.ERROR)
```

```python
print(DEFAULT_DEVICE)
try:
    i = 0
    with open(CONFIG["preprocess"]["new_dataset_index"], 'a') as file:
        for path in Path(CONFIG["preprocess"]["dataset"]).rglob('*.mid'):
            name, _ = splitext(basename(path))
            enc = midichar.encode_midi(
                midi_file=str(path),
                window_size=CONFIG["preprocess"]["window_size"],
                instrument_name="Standard Kit")

            if enc is not None:
                i += 1
                midichar.decode_midi(notes=enc,
                        out_file=f"drums/{i}.mid",
                        instrument_name="Standard Kit")
                # file.write(str(enc.tolist())+"\n")
                file.write(str(f"drums/{i}.mid")+"\n")
            if i == CONFIG["preprocess"]["max_dataset_items"]:
                break
except Exception as e:
    print(f"An error occurred: {e}")
```
